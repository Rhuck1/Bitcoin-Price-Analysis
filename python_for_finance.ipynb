Implementation of short and long term rolling averages on data

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data
plt.style.use('ggplot')
```

```python
# Defunen the instruments to download. We Would like to see Apple, Microsoft and the S&P500 index
tickers = ['AAPL', 'MSFT', 'SPY', '^GSPC']

# We would like all available data form 01/01/2000 until 12/31/2016
start_date = '2010-01-01'
end_date = '2016-12-31'

# Use pandas_reader.data.Datareader to load the desired data
# panel_data = data.DataReader('INPX', data_source='yahoo', start=start_date, end=end_date)
panel_data = yf.download(tickers=tickers, start=start_date, end=end_date)
```

```python
# Getting 
close = panel_data['Close'].copy()
```

```python
close.describe()
```

```python
# Get the MSFT timeseries
msft = close.loc[:, 'MSFT']

# Calculate the 20 and 100 days moving average fo the closing prices
short_rolling_msft = msft.rolling(window=20).mean()
long_rolling_msft = msft.rolling(window=100).mean()

# Plot msft
fig, ax = plt.subplots(figsize=(16, 9))

ax.plot(msft.index, msft, label='MSFT', color='black')
ax.plot(short_rolling_msft.index, short_rolling_msft, label='20 days rolling', color='red')
ax.plot(long_rolling_msft.index, long_rolling_msft, label='100 days rolling', color='blue')

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted Closing Price USD ($)')
ax.legend();
```
